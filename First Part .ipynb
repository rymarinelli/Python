```python
data = pd.read_csv("MovieData.csv", sep='\t')
```

```python
data.head()
```

```python
data = data.dropna()
data.head()
```

```python
data.columns = data.columns.str.strip().str.replace(" ","_")
data.columns
```

```python
budget = data.Budget.to_frame()
newData = data[budget.Budget > 50000000]
newData = newData[newData.US_Gross != "Unknown"]
newData.US_Gross.median()
```

```python
#distributor = data.Distributor.unique()
#distSet = set(distributor)
instances = {}
for a in data.Distributor:
    if a not in instances.keys():
        instances[a] = 0;
    instances[a] += 1
    print(instances)
```

```python
with open("MovieCount.csv","w", newline='') as f:
    column_names = ['distributor', 'film_count']
    dictWriter = csv.DictWriter(f, column_names)
    dictWriter.writeheader()
      
    for k, v in instances.items():
        dictWriter.writerow({'distributor': k , 'film_count' : v})
```

```python
test = data.Release_Date

testList = []
testListTwo = []
testListThree = []

for i in test:
    print(i)
    print(i[6])
    if i[6] == "9":
      	testList.append(i)
    elif i[6] == "0":
        testListTwo.append(i)
    elif i[6] == "1":
        testListThree.append(i)
```

```python
testDF = pd.DataFrame(testList, columns = ['Release_Date'])
testDFTwO = pd.DataFrame(testListTwo, columns = ['Release_Date'])
testDFThree = pd.DataFrame(testListThree, columns = ['Release_Date'])
```

```python
testOne = pd.merge(testDF, data , on= 'Release_Date', how= 'left')
testTwo = pd.merge(testDFTwO, data , on= 'Release_Date', how= 'left')
testThree = pd.merge(testDFThree, data , on= 'Release_Date', how= 'left')
```

```python
testTwo = testTwo[testTwo.US_Gross != "Unknown"]
testThree = testThree[testThree.US_Gross != "Unknown"]
testTwo.US_Gross.median()  < testOne.US_Gross.median()
testThree.US_Gross.median() > testOne.US_Gross.median()
print(testOne.US_Gross.median())
print(testTwo.US_Gross.median())
print(testThree.US_Gross.median())
```
